##**Caso de Estudio**
Se busca implementar un modelo de árbol de decisión que permita pronosticar si una persona desarrollará enfermedad cardíaca o posee un alto riesgo cardiovascular, a partir de variables clínicas como:

* Edad (Age): que es son los años de vida de la persona al momento de la evaluación.

* Presión sistólica (Systolic): Es la fuerza generada cada vez que el corazón late y bombea sangre a las arterias, y mide la máxima presión cuando el corazón se contrae.

* Presión diastólica (Diastolic): mide la presión mínima cuando el corazón se relaja entre latidos.

* Índice de masa corporal (BMI): , una herramienta médica estándar que evalúa si una persona tiene un peso saludable en relación con su estatura. Se calcula dividiendo el peso en kilogramos por el cuadrado de la altura en metros, para clasificar el bajo peso, peso normal, sobrepeso u obesidad.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
```

1. Se procede con la carga de los datos de trabajo

```python
nxl = '/content/sample_data/2. HeartDisease.xlsx'
XDB = pd.read_excel(nxl, sheet_name=0)

XD = XDB[['Age','Systolic','Diastolic','BMI']]
yd = XDB['Disease']
```

2. Se procede con la implementación del modelo de árbol

```python
mar = DecisionTreeClassifier(criterion='gini', max_depth=4)
mar.fit(XD, yd)

ydp = mar.predict(XD)

cm = confusion_matrix(yd, ydp)

VN = cm[0,0]
FP = cm[0,1]
FN = cm[1,0]
VP = cm[1,1]

Exactitud = (VP + VN) / len(XD)
print("La exactitud es: ", Exactitud)

Sensibilidad = VP / (VP + FN)
print("La sensibilidad es: ", Sensibilidad)

Especificidad = VN / (VN + FP)
print("La especificidad es: ", Especificidad)

Precision = VP / (VP + FP)
print("La precisión es: ", Precision)

PrecisionNegativa = VN / (VN + FN)
print("La precisión negativa es: ", PrecisionNegativa)
```

3. Despliegue del Árbol de Decisión

```python
from sklearn.tree import export_graphviz
from pydotplus import graph_from_dot_data

vs = ["Age", "Systolic", "Diastolic", "BMI"]

dot_data=export_graphviz(mar, feature_names=vs)
graph=graph_from_dot_data(dot_data)
graph.write_png('Arbol.png')
```
